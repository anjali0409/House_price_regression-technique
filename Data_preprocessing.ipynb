```python
#fill null value with mode of the class
df[i].fillna(df[i].value_counts().index[0], inplace=True)
df.isna().sum()
```

## Determine the character based columns

```python

cat=[]  #character based column
for i in df.columns:
    if i not in num:
        cat.append(i)
len(cat),len(num),len(df.columns)
```

```python
cat_cont=[]
cat_class=[]
for i in cat:
    if df[i].nunique()>8:
        cat_cont.append(i)
    else:
        cat_class.append(i)
len(cat_cont),len(cat_class)
print(cat_cont)
print(cat_class)
```

## Drop the cat_cont columns

```python
df.drop(cat_cont, axis=1, inplace=True)
df.isna().sum()
```

```python
#fill null vlaue in character column with mode(more unique values)
for i in cat_class:
    df[i].fillna(df[i].value_counts().mode, inplace=True)
df.isna().sum() 
```

```python
#fill null vlaue in character column with median(less unique values)
for i in cat_class:
    df[i].fillna(df[i].value_counts().median, inplace=True)
df.isna().sum()
```

# To determine outliers columns

Outliers in input data can skew and mislead the training process of machine learning algorithms resulting in longer training times, less accurate models and ultimately poorer results.

```python
outliers_columns=[]
for i in df.describe().columns:
    if df[i].nunique()>8:
        outliers_columns.append(i)
```

```python
outliers_columns
```

```python
for i in outliers_columns:
    x=df[i].values
    import numpy as np
    qrt_1=np.quantile(x,0.25)
    qrt_3=np.quantile(x,0.75)
    qrt_1
    qrt_3
    iqr=qrt_3-qrt_1
    utv=qrt_3 + 1.5*iqr
    ltv=qrt_1 - 1.5*iqr
    print(iqr)
```

```python
outliers=[]
for i in x:
    if i < ltv or i>utv:
        outliers.append(i)
outliers=np.array(outliers)
outliers.min()
outliers.max()
outliers_treatment=[]
for i in x:
    if i < ltv or i>utv:
        outliers_treatment.append(np.median(x))
    else:
        outliers_treatment.append(i)
df[i]=outliers_treatment
df[i]  
```
